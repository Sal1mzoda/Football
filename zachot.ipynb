```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df = pd.read_csv('Football_English.csv')
```

```python
# Построение гистограммы количества матчей по дням недели
plt.figure(figsize=(10, 6))
sns.countplot(x='Date.day-of-week', data=df)
plt.title('Распределение матчей по дням недели')
plt.xlabel('День недели (0 - Понедельник, 6 - Воскресенье)')
plt.ylabel('Количество матчей')
plt.xticks(range(7), ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс'])
plt.show()
```

```python
home_wins = df[df['Winner'] == 'Home'].shape[0]
away_wins = df[df['Winner'] == 'Away'].shape[0]
draws = df[df['Winner'] == 'Draw'].shape[0]

print(f'Home Wins: {home_wins}, Away Wins: {away_wins}, Draws: {draws}')
```

```python
# Группировка данных по месяцам
# Группировка данных по месяцам
monthly_home_goals = df.groupby('Date.month')['Home goals'].mean().reset_index()
monthly_away_goals = df.groupby('Date.month')['Away goals'].mean().reset_index()

# Построение графика изменения среднего количества голов в течение месяца
plt.figure(figsize=(12, 6))
plt.plot(monthly_home_goals['Date.month'], monthly_home_goals['Home goals'], label='Домашние голы', marker='o')
plt.plot(monthly_away_goals['Date.month'], monthly_away_goals['Away goals'], label='Гостевые голы', marker='x')
plt.title('Среднее количество голов в течение месяца')
plt.xlabel('Месяц')
plt.ylabel('Среднее количество голов')
plt.legend()
plt.xticks(range(1, 13), ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн', 'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек'])
plt.grid(True)
plt.show()
```

```python
average_home_goals = df['Home goals'].mean()
average_away_goals = df['Away goals'].mean()

print(f'Average Home Goals: {average_home_goals}, Average Away Goals: {average_away_goals}')
```

```python
# Гистограмма голов
plt.figure(figsize=(10, 6))
sns.histplot(df['Home goals'], kde=True, color='blue', label='Home Goals')
sns.histplot(df['Away goals'], kde=True, color='red', label='Away Goals')
plt.legend()
plt.title('Распределение голов')
plt.show()
```

```python
average_home_yellow_cards = df['Home yellow cards'].mean()
average_away_yellow_cards = df['Away yellow cards'].mean()
average_home_red_cards = df['Home red cards'].mean()
average_away_red_cards = df['Away red cards'].mean()

print(f'Average Home Yellow Cards: {average_home_yellow_cards}, Average Away Yellow Cards: {average_away_yellow_cards}')
print(f'Average Home Red Cards: {average_home_red_cards}, Average Away Red Cards: {average_away_red_cards}')
```

```python
# Построение гистограммы количества матчей по месяцам
plt.figure(figsize=(10, 6))
sns.countplot(x='Date.month', data=df)
plt.title('Распределение матчей по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество матчей')
plt.show()

```

```python
average_home_turnover = df['Home Turnover (£m)'].mean()
average_away_turnover = df['Away Turnover (£m)'].mean()

print(f'Average Home Turnover: {average_home_turnover}, Average Away Turnover: {average_away_turnover}')
```

```python
# Выбор числовых переменных
numerical_columns = [
    'Home goals', 'Away goals', 'Half time home goals', 'Half time away goals', 
    'Home shots', 'Away shots', 'Home shots on target', 'Away shots on target', 
    'Home fouls commited', 'Away fouls commited', 'Home corners', 'Away corners', 
    'Home yellow cards', 'Away yellow cards', 'Home red cards', 'Away red cards', 
    'Home Turnover (£m)', 'Home Gate and Matchday income (£m)', 'Home TV and Broadcasting (£m)', 
    'Home Commercial income (£m)', 'Away Turnover (£m)', 'Away Gate and Matchday income (£m)', 
    'Away TV and Broadcasting (£m)', 'Away Commercial income (£m)'
]

# Построение матрицы корреляции
correlation_matrix = df[numerical_columns].corr()

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Матрица корреляции числовых переменных')
plt.show()
```

```python
# Выбор переменных, связанных с доходами и результатами матчей
selected_columns = [
    'Home goals', 'Away goals', 'Home Turnover (£m)', 
    'Home Gate and Matchday income (£m)', 'Home TV and Broadcasting (£m)', 
    'Home Commercial income (£m)', 'Away Turnover (£m)', 
    'Away Gate and Matchday income (£m)', 'Away TV and Broadcasting (£m)', 
    'Away Commercial income (£m)'
]

# Построение матрицы корреляции для выбранных переменных
correlation_matrix = df[selected_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Матрица корреляции доходов и результатов матчей')
plt.show()

```

```python
# Построение сводной таблицы для анализа влияния статуса защиты профиля на количество забитых голов домашней командой
# Построение сводной таблицы для анализа влияния статуса защиты профиля на количество забитых голов домашней командой
protected_goals = df.groupby('Home team')['Home goals'].mean()

# Визуализация
plt.figure(figsize=(10, 6))
sns.barplot(x=protected_goals.index, y=protected_goals.values)
plt.title('Среднее количество забитых голов домашней командой')
plt.xlabel('Команда')
plt.ylabel('Среднее количество забитых голов')
plt.xticks(rotation=90)
plt.show()
```

```python
# Группировка по дню недели проведения матча и вычисление среднего количества голов, забитых домашней командой
home_goals_by_dayofweek = df.groupby('Date.day-of-week')['Home goals'].mean()

# Визуализация
plt.figure(figsize=(10, 6))
sns.barplot(x=home_goals_by_dayofweek.index, y=home_goals_by_dayofweek.values, palette='magma')
plt.title('Среднее количество голов, забитых домашней командой, по дням недели')
plt.xlabel('День недели')
plt.ylabel('Среднее количество голов')
plt.xticks(range(7), ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс'])
plt.show()
```
